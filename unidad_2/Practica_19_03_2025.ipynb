```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

#definicion de colores

#colores=[(0,'black'),(1,'white')] #para genera escala de grises
colores=[(0,'black'),(0.01,'purple'),(0.5,'cyan'),(1,'yellow')] #para genera escala de grises

cmap= LinearSegmentedColormap.from_list('degradado',colores) #generamos a travez de nuestro array
imagen_degra=cmap(np.linspace(0,1,100).reshape(-1,1))

plt.imshow(imagen_degra,aspect='auto')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np



plt.rcParams['figure.figsize']=[10,5]
tmanio_imagen=800
R=np.ones((tmanio_imagen,tmanio_imagen))*0
G=np.ones((tmanio_imagen,tmanio_imagen))*0
B=np.ones((tmanio_imagen,tmanio_imagen))*0

imagen=np.dstack((R,G,B))
imagen=np.array(imagen,dtype=int)
white=[255,255,255]

#imagen[0:50,0:50]=(white)
#imagen[100:150,0:50]=(white)


for i in range(0,800,200):
    imagen[i:i+100,0:100]=white

for i in range(100,800,200):
    imagen[i:i+100,100:200]=white
    
for i in range(0,800,200):
    imagen[i:i+100,200:300]=white

for i in range(100,800,200):
    imagen[i:i+100,300:400]=white

for i in range(0,800,200):
    imagen[i:i+100,400:500]=white
    
for i in range(100,800,200):
    imagen[i:i+100,500:600]=white

for i in range(0,800,200):
    imagen[i:i+100,600:700]=white

for i in range(100,800,200):
    imagen[i:i+100,700:800]=white

plt.figure()

plt.imshow(imagen)
plt.show
```
