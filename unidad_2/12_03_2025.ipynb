Umbralizaciones por defecto tuzero en opneCv

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

path="../imagenes/pardoOriginal.png"
imagen=cv2.imread(path,0)
_,umbr_trunc=cv2.threshold(imagen,127,255,cv2.THRESH_TRUNC)
_,umbr_tozero=cv2.threshold(imagen,127,255,cv2.THRESH_TOZERO)
_,umbr_tozero_inv=cv2.threshold(imagen,127,255,cv2.THRESH_TOZERO_INV)

fg,axs=plt.subplots(1,4,figsize=(12,5))

axs[0].imshow(imagen,cmap='gray')
axs[0].set_title('Imagen Original')

axs[1].imshow(umbr_trunc,cmap='gray')
axs[1].set_title('Imagen trunc')

axs[2].imshow(umbr_tozero,cmap='gray')
axs[2].set_title('Imagen tozero')

axs[3].imshow(umbr_tozero_inv,cmap='gray')
axs[3].set_title('Imagen tozero_INV')

```

Contraste

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def contraste(img,factor=1.0):
    imagen=img.astype(np.float32)
    media=np.mean(imagen)
    img_contraste=media+(imagen-media)*factor
    img_contraste=np.clip(img_contraste,0,255).astype(np.uint8)
    return img_contraste

img=cv2.imread(path,0)
factor_contraste=4
imagen_neva= contraste(img,factor_contraste)

fg,axs=plt.subplots(1,2,figsize=(12,5))

axs[0].imshow(img,cmap='gray')
axs[0].set_title('Imagen Original')

axs[1].imshow(imagen_neva,cmap='gray')
axs[1].set_title('Imagen con contraste')
```

Contraste  a colores


```python
def ContrasteColor(imagen,factor=1.0):
    img=imagen.astype(np.float32)
    for i in range(3):
        media = np.mean(img[:,:,i])
        img[:,:,i]=media+(img[:,:,i]-media)*factor
        img[:,:,i]=np.clip(img[:,:,i],0,255)
        return img.astype(np.uint8)


img=cv2.imread(path)
factor=4
img_nueva=ContrasteColor(img,factor)


fg,axs=plt.subplots(1,2,figsize=(12,5))

axs[0].imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
axs[0].set_title('Imagen Original')

axs[1].imshow(cv2.cvtColor(img_nueva,cv2.COLOR_BGR2RGB))
axs[1].set_title('Imagen contraste color')

```

histograma de una imagen constrastada


```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

def contraste(imagen,factor=1.0):
    imagen=imagen.astype(np.float32)
    media=np.mean(imagen)
    img_contrastada=media+(imagen-media)*factor
    img_contrastada=np.clip(img_contrastada,0,255).astype(np.uint8)
    return img_contrastada

img=cv2.imread(path,0)

f_contraste=6
img_nueva= contraste(img,f_contraste)

histograma_ori=cv2.calcHist([img],[0],None,[256],[0,256])
histograma_nuevo_contrans=cv2.calcHist([img_nueva],[0],None,[256],[0,256])

plt.figure(figsize=(10,5))

plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')

plt.subplot(2,2,2)
plt.plot(histograma_ori,color='gray')
plt.xlabel('intencidad')
plt.ylabel('frecuencia')


plt.subplot(2,2,3)
plt.imshow(img_nueva,cmap='gray')

plt.subplot(2,2,4)
plt.plot(histograma_nuevo_contrans,color='gray')
plt.xlabel('intencidad')
plt.ylabel('frecuencia')

plt.show()
```

mascaras

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

path='../imagenes/pardoOriginal.png'
img=cv2.imread(path,0)
#dimenciones de imagen
alto,ancho=img.shape[:2]
#crear mascara circular
mascara=np.zeros_like(img)
centro=(ancho//2 ,alto//2)
radio=100
cv2.circle(mascara,centro,radio,(255),-1)
img_mascara=cv2.bitwise_and(img,img,mask=mascara)

fg,axs=plt.subplots(1,2,figsize=(12,5))

axs[0].imshow(img,cmap='gray')
axs[0].set_title('Imagen Original')

axs[1].imshow(img_mascara,cmap='gray')
axs[1].set_title('Imagen con mascara')
```
