UMBRALIZACION CON OTSU

La forma en la que trabaja esta tecnica es que dimenciona una imagen como un array de de N pixeles y lo que hace es de la forma mas optima dividir estos pixeles en dos grupos.
el primero desde 1 hasta t y el otro de t+1 hasta L-1 esto lo hace con una imagen que este en escala de grises y aplica las formulas de:

UMBRAL_BINARIO dst(x,y)={ maxValue        if src(x,y)  >  T(x,y)  }
                        {    0                  otherwise         }



UMBRAL_INV_BINARIO dst(x,y) {      0               if src(x,y) > T(x,y)  }  
                            {  maxValue            otherwise             }

esta funcion es de las mas optimas y utilies para todo tipo de imagenes pues a comparacion de otros metodos funciona de mejor manera ante la falta de luz o ruido en la imagen es decir es mas utili en casos reales.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

path="../imagenes/polarOriginal.jpg"

imagen=cv2.imread(path,0)
_,imagen_otsu=cv2.threshold(imagen,0,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)

umbral_imagen = np. where(imagen>imagen_otsu,255,0).astype(np.uint8)

fig,axs = plt.subplots(1,2,figsize=(20,5))

axs[0].imshow(imagen,cmap='gray')
axs[0].set_title('imagen Original')

axs[1].imshow(imagen_otsu,cmap='gray')
axs[1].set_title('imagen con umbral de otsu')
```
