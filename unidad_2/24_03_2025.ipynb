Ecualizacion de histograma - se desea normalizar una imagen a un punto optimo

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

path="../imagenes/pardoOriginal.png"
imagen=cv2.imread(path,0)
histograma,calc=np.histogram(imagen.flatten(),256,[0,256])

distAcum=histograma.cumsum()
distAcum_nomalizado=distAcum*histograma.max()/distAcum.max()
ecualizado=cv2.equalizeHist(imagen)
histograma_equa,_=np.histogram(ecualizado.flatten(),256,[0,256])

plt.figure(figsize=(10,5))

plt.subplot(2,2,1)
plt.imshow(imagen,cmap='gray')
plt.title("Original")

plt.subplot(2,2,2)
plt.plot(distAcum,color='b')
plt.hist(imagen.flatten(),256,[0,256],color='r')
plt.xlim([0,256])
plt.legend(('Distribucion','Histrograma'),loc='upper left')
plt.title("Histograma original")

plt.subplot(2,2,3)
plt.imshow(ecualizado,cmap='gray')
plt.title("Imagen ecualizada")

plt.subplot(2,2,4)
plt.plot(histograma_equa,color='b')
plt.xlim([0,256])
plt.tight_layout()
plt.show()




```

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt

imagen = cv2.imread(path)
imagen_gray=cv2.imread(path,0)

ecualizada=cv2.equalizeHist(imagen_gray)
ecu_color=cv2.merge([cv2.equalizeHist(imagen[:,:,0]),cv2.equalizeHist(imagen[:,:,1]),cv2.equalizeHist(imagen[:,:,2])])

plt.imshow(ecualizada,cmap='gray')
plt.show()

plt.imshow(cv2.cvtColor(ecu_color,cv2.COLOR_BGR2RGB))
plt.show()
```

clehe

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt

imagen = cv2.imread(path,0)

clip=2.0
tiles=8
clahe=cv2.createCLAHE(clipLimit=clip,tileGridSize=(tiles,tiles))
ecualizado=clahe.apply(imagen)

plt.imshow(imagen,cmap='gray')
plt.show()

plt.imshow(ecualizado,cmap='gray')
plt.show()
```

transformada de forier

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt

imagen = cv2.imread(path,0)

transformada=np.fft.fft2(imagen) #aplicando transfomada e inversa de forier
desplazar=np.fft.fftshift(transformada) # separo el espectro
magnitu_espectro=np.log(1+np.abs(desplazar))

plt.imshow(imagen,cmap='gray')
plt.show()

plt.imshow(magnitu_espectro,cmap='gray')
plt.show()
```
