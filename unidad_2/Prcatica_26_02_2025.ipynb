El filtro bayer es usado en practicamente todos los dispositivos que poseen una camara pues este tipo de filtro es que las camaras dedican  4 forodetectores los cuales se encargan de separar los colores de las imagenes que capturan, estas imagenes estan conformadas por los ya muy conocidos colores (RGB) y la distribucion de como los 4 fotodetectors administran es 2 de ellos toman el verde y uno al rojo y otro al azul 
la razcon por la cual se le da mas cantidad al verde es que el ojo humando es mas sensible a la persepcion de ese color
una vez tomando ya el dispositivo sr encarga de poder unificar estos colores y mostrar el resultado.
Ahora nombraremos las distitas versiones que posee este filtro :

cv2.COLOR_BAYER_BG2BGR: Convierte una imagen Bayer BG a BGR.
cv2.COLOR_BAYER_GB2BGR: Convierte una imagen Bayer GB a BGR.
cv2.COLOR_BAYER_RG2BGR: Convierte una imagen Bayer RG a BGR.
cv2.COLOR_BAYER_GR2BGR: Convierte una imagen Bayer GR a BGR.
cv2.COLOR_BAYER_BG2RGB: Convierte una imagen Bayer BG a RGB.
cv2.COLOR_BAYER_GB2RGB: Convierte una imagen Bayer GB a RGB.
cv2.COLOR_BAYER_RG2RGB: Convierte una imagen Bayer RG a RGB.
cv2.COLOR_BAYER_GR2RGB: Convierte una imagen Bayer GR a RGB.
cv2.COLOR_BAYER_BG2GRAY: Convierte una imagen Bayer BG a escala de grises.
cv2.COLOR_BAYER_GB2GRAY: Convierte una imagen Bayer GB a escala de grises.
cv2.COLOR_BAYER_RG2GRAY: Convierte una imagen Bayer RG a escala de grises.
cv2.COLOR_BAYER_GR2GRAY: Convierte una imagen Bayer GR a escala de grises.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

path="../imagenes/pardoOriginal.png"
img= cv2.cvtColor(cv2.imread(path,cv2.IMREAD_COLOR),cv2.COLOR_BGR2RGB)
bayer=cv2.cvtColor(path,cv2.COLOR_BayerRGB2BG)

cv2.imwrite('../img_modificadas/img_filtro_bayer_26_02_2025.jpg', bayer)

fig,axs=plt.subplots(1,2,figsize=(20,10))

axs[0].imshow(img)
axs[0].set_title("original")

axs[1].imshow(bayer)
axs[1].set_title("filtro bayer")
```

```python
import numpy as np

def RGB2YUV( rgb ):
     
    m = np.array([[ 0.29900, -0.16874,  0.50000],
                 [0.58700, -0.33126, -0.41869],
                 [ 0.11400, 0.50000, -0.08131]])
     
    yuv = np.dot(rgb,m)
    yuv[:,:,1:]+=128.0
    return yuv

def YUV2RGB( yuv ):
      
    m = np.array([[ 1.0, 1.0, 1.0],
                 [-0.000007154783816076815, -0.3441331386566162, 1.7720025777816772],
                 [ 1.4019975662231445, -0.7141380310058594 , 0.00001542569043522235] ])
    
    rgb = np.dot(yuv,m)
    rgb[:,:,0]-=179.45477266423404
    rgb[:,:,1]+=135.45870971679688
    rgb[:,:,2]-=226.8183044444304
    return rgb
```

```python
path="../imagenes/pardoOriginal.png"
img= cv2.cvtColor(cv2.imread(path,cv2.IMREAD_COLOR),cv2.COLOR_BGR2RGB)
img_yuv=RGB2YUV(img)
img_rgb=YUV2RGB(img_yuv)

fig,axs=plt.subplots(1,3,figsize=(20,10))

axs[0].imshow(img)
axs[0].set_title("original")

axs[1].imshow(img_yuv)
axs[1].set_title("convercion RGB_YUV")

axs[2].imshow(img_rgb)
axs[2].set_title("convercion YUV_RGB")
```
