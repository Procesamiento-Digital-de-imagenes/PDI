```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def solarizada(imagen,umbral):
    imagen_solarizada=np.where(imagen<umbral,imagen,255-imagen)
    return imagen_solarizada


path="../imagenes/pardoOriginal.png"
imagen =cv2.imread(path)
img_gris=cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
umbral=127

imagen_nueva=solarizada(imagen,umbral)

fig,axs=plt.subplots(1,3,figsize=(20,10))
axs[0].imshow(cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB))
axs[0].set_title("Imagen original")

axs[1].imshow(img_gris)
axs[1].set_title("Imagen gris")

axs[2].imshow(imagen_nueva,cmap='gray')
axs[2].set_title("Imagen solarizada")

```

Posterizasa

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def porterizada(imagen,nivel):
    factor=255/nivel
    imagen_porst=np.round((imagen/factor)*factor)
    return imagen_porst.astype(np.uint8)

imagen = cv2.imread(path)
img_gris=cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)

nivel=4 # nivel de intensidad
nueva_imagen= porterizada(img_gris,nivel)

fig,axs=plt.subplots(1,3,figsize=(20,10))

axs[0].imshow(cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB))
axs[0].set_title("Imagen original")

axs[1].imshow(imagen_nueva,cmap='gray')
axs[1].set_title("Imagen posterizada")


```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

array=[
    {"name": "COLORMAP_AUTUMN", "value": cv2.COLORMAP_AUTUMN},
    {"name": "COLORMAP_BONE", "value": cv2.COLORMAP_BONE},
    {"name": "COLORMAP_JET", "value": cv2.COLORMAP_JET},
    {"name": "COLORMAP_WINTER", "value": cv2.COLORMAP_WINTER},
    {"name": "COLORMAP_RAINBOW", "value": cv2.COLORMAP_RAINBOW},
    {"name": "COLORMAP_OCEAN", "value": cv2.COLORMAP_OCEAN},
    {"name": "COLORMAP_SUMMER", "value": cv2.COLORMAP_SUMMER},
    {"name": "COLORMAP_SPRING", "value": cv2.COLORMAP_SPRING},
    {"name": "COLORMAP_COOL", "value": cv2.COLORMAP_COOL},
    {"name": "COLORMAP_HSV", "value": cv2.COLORMAP_HSV},
    {"name": "COLORMAP_PINK", "value": cv2.COLORMAP_PINK},
    {"name": "COLORMAP_HOT", "value": cv2.COLORMAP_HOT},
    {"name": "COLORMAP_PARULA", "value": cv2.COLORMAP_PARULA},
    {"name": "COLORMAP_MAGMA", "value": cv2.COLORMAP_MAGMA},
    {"name": "COLORMAP_INFERNO", "value": cv2.COLORMAP_INFERNO},
    {"name": "COLORMAP_PLASMA", "value": cv2.COLORMAP_PLASMA},
    {"name": "COLORMAP_VIRIDIS", "value": cv2.COLORMAP_VIRIDIS},
    {"name": "COLORMAP_CIVIDIS", "value": cv2.COLORMAP_CIVIDIS},
    {"name": "COLORMAP_TWILIGHT", "value": cv2.COLORMAP_TWILIGHT},
    {"name": "COLORMAP_TWILIGHT_SHIFTED", "value": cv2.COLORMAP_TWILIGHT_SHIFTED},
    {"name": "COLORMAP_TURBO", "value": cv2.COLORMAP_TURBO},
    {"name": "COLORMAP_DEEPGREEN", "value": cv2.COLORMAP_DEEPGREEN}
]

def Falso_color(imagen_gray,valor):
    escalar_imagen=cv2.normalize(imagen_gray,None,0,255,cv2.NORM_MINMAX)
    apli_color=cv2.applyColorMap(escalar_imagen,array[valor]['value'])
    return apli_color

image_gray=cv2.imread(path,0)

fig,axs=plt.subplots(22,1,figsize=(100,50))

#axs[0].imshow(cv2.cvtColor(image_gray,cv2.COLOR_BGR2RGB))
#axs[0].set_title("Imagen original")
for i in range(len(array)):
    img_falso=Falso_color(image_gray,i)
    axs[i].imshow(cv2.cvtColor(img_falso,cv2.COLOR_BGR2RGB))
    axs[i].set_title(array[i]['name'])
```
