CREACION DE UNA IMAGEN


```python
import matplotlib.pyplot as plt
import cv2
import numpy as np


plt.rcParams['figure.figsize']=[10,5]
tmanio_imagen=100
R=np.ones((tmanio_imagen,tmanio_imagen))*0
G=np.ones((tmanio_imagen,tmanio_imagen))*255
B=np.ones((tmanio_imagen,tmanio_imagen))*0

imagen=np.dstack((R,G,B))
imagen=np.array(imagen,dtype=int)

imagen[0:50,0:50]=(255,0,0)
imagen[50:100,0:50]=(0,0,0)
imagen[0:50,50:100]=(0,0,255)
imagen[0:50,0:50]=(255,0,0)
plt.figure()

plt.imshow(imagen)
plt.show

cv2.imwrite('../img_modificadas/img_Cuadrantes_24_02_2025.jpg',imagen)
```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

path="../imagenes/pardoOriginal.png"
fig,axs =plt.subplots(1,2,figsize=(25,3))

imagen=cv2.imread(path,cv2.IMREAD_GRAYSCALE)
img_inicial = imagen.dtype #valor profundidad del color
print('profuncidad del color:',img_inicial)

img_16bits=np.uint16(imagen)
img_nueva = img_16bits.dtype
print('profuncidad del color:',img_nueva)

axs[0].imshow(imagen)
axs[0].set_title("imagen")

axs[1].imshow(img_16bits)
axs[1].set_title("imagen")

#cv2.imwrite('../img_modificadas/img_24_02_2025.jpg',img_16bits)

```

RGB a HSV

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
fig,axs =plt.subplots(1,4,figsize=(25,3))
path="../imagenes/pardoOriginal.png"

img_bgr=img=cv2.imread(path)
img_rgb=cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
hsv=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HSV)

axs[0].imshow(img_rgb)
axs[0].set_title("imagen")

fig_img1=axs[1].imshow(hsv[...,0],cmap='gray',vmin=0,vmax=179)
fig.colorbar(fig_img1,ax=axs[1],shrink=0.3)
axs[1].set_title("canal H")

fig_img2=axs[2].imshow(hsv[...,1],cmap='gray',vmin=0,vmax=255)
fig.colorbar(fig_img2,ax=axs[2],shrink=0.3)
axs[2].set_title("canal H")

fig_img3=axs[3].imshow(hsv[...,2],cmap='gray',vmin=0,vmax=255)
fig.colorbar(fig_img3,ax=axs[3],shrink=0.3)
axs[3].set_title("canal H")

```
