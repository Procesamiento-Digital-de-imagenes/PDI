SOLARIZACION NEGATIVA

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def solarizada(imagen,umbral):
    imagen_solarizada=np.where(imagen<umbral,imagen,255-imagen)
    return imagen_solarizada

path="../imagenes/pardoOriginal.png"
imagen =cv2.imread(path)
img_solarizada=solarizada(imagen,0)

imagen_negativo=255-img_solarizada

plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB))

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(img_solarizada,cv2.COLOR_BGR2RGB))

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(imagen_negativo,cv2.COLOR_BGR2RGB))

plt.show()
```
