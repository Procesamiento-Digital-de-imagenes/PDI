```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
 
path="./Examen.jpg"
imagen=cv2.imread(path)
imagen_rgb= cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
img_cortada=imagen_rgb[1:200,1:200,:]

plt.subplot(2,2,1)
plt.imshow(img_cortada)

plt.subplot(2,2,2)
color=('r','g','b')
for canal,col in enumerate(color):
    histograma=cv2.calcHist([img_cortada*255],[canal],None,[256],[0,256])
    plt.plot(histograma,color=col)
    plt.xlim([0,256])
plt.show


#solarizacion
def solarizada(imagen,umbral):
    imagen_solarizada=np.where(imagen<umbral,imagen,255-imagen)
    return imagen_solarizada

imagen =cv2.imread(path)
img_gris=cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
umbral=127

imagen_solarizada=solarizada(imagen,umbral)

fig,axs=plt.subplots(1,2,figsize=(20,10))
axs[0].imshow(cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB))
axs[0].set_title("Imagen original")

axs[1].imshow(imagen_solarizada,cmap='gray')
axs[1].set_title("Imagen solarizada")

#posterizada
def porterizada(imagen,nivel):
    factor=255/nivel
    imagen_porst=np.round((imagen/factor)*factor)
    return imagen_porst.astype(np.uint8)

imagen = cv2.imread(path)
img_gris=cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)

nivel=4 # nivel de intensidad
imagen_posterizada= porterizada(img_gris,nivel)

fig,axs=plt.subplots(1,2,figsize=(20,10))

axs[0].imshow(cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB))
axs[0].set_title("Imagen original")

axs[1].imshow(imagen_posterizada,cmap='gray')
axs[1].set_title("Imagen posterizada")


#umbralizacion
def dos_picos(img):
    hist=cv2.calcHist([imagen],[0],None,[256],[0,256])
    pico=np.array(hist)
    pico=pico.flatten()
    pico_dif=np.diff(pico)
    pico_mayor=np.argmax(pico_dif[:128])
    pico_menor=np.argmax(pico_dif[128:])+128

    umbral_2picos=int((pico_mayor+pico_menor)/2)
    return umbral_2picos


imagen=cv2.imread(path,0)

umbral_2pic=dos_picos(imagen)

_,imagen_2picos=cv2.threshold(imagen,umbral_2pic,255,cv2.THRESH_BINARY)

fig,axs=plt.subplots(1,2,figsize=(10,5))

axs[0].imshow(imagen,cmap='gray')
axs[0].set_title("imagen original")

axs[1].imshow(imagen_2picos,cmap='gray')
axs[1].set_title("imagen dos picos")


plt.imsave('./img_cortada.jpg',img_cortada)
plt.imsave('./img_solarizada.jpg',imagen_solarizada)
plt.imsave('./img_posterizada.jpg',imagen_posterizada)
plt.imsave('./img_imagen_2picos.jpg',imagen_2picos)






```
