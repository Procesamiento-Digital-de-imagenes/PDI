Erosion

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('../imagenes/pardoOriginal.png',cv2.IMREAD_GRAYSCALE)
#ELEMNETO ESTRUCTURANTE
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
#aplicar erosion
img_rest =cv2.erode(imagen,kernel)
plt.subplot(121)
plt.imshow(imagen,cmap='gray')

plt.subplot(122)
plt.imshow(img_rest,cmap='gray')
```

erosion binaria

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('../imagenes/pardoOriginal.png',cv2.IMREAD_GRAYSCALE)
_,img_bin=cv2.threshold(imagen,127,255,cv2.THRESH_BINARY_INV)
#ELEMNETO ESTRUCTURANTE
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
#aplicar erosion
img_rest =cv2.erode(img_bin,kernel)#dilate
plt.subplot(121)
plt.imshow(imagen,cmap='gray')

plt.subplot(122)
plt.imshow(img_rest,cmap='gray')
```

apertura(erosion dilatasion)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('../imagenes/pardoOriginal.png',cv2.IMREAD_GRAYSCALE)
_,img_bin=cv2.threshold(imagen,127,255,cv2.THRESH_BINARY_INV)
#ELEMNETO ESTRUCTURANTE
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(9,9))
#aplicar erosion
#img_rest =cv2.erode(img_bin,kernel)#dilate
apertura=cv2.morphologyEx(imagen,cv2.MORPH_OPEN,kernel)
plt.subplot(121)
plt.imshow(imagen,cmap='gray')

plt.subplot(122)
plt.imshow(apertura,cmap='gray')
```

SKIMAGE CON EROSION DILATACION APERTURA Y CERRADURA

```python
from skimage import data , morphology
import matplotlib.pyplot as plt

image =  data.binary_blobs(length=64,n_dim=2)
kernel = morphology.disk(5)
apertura=morphology.binary_opening(image,kernel)
cerradura=morphology.binary_closing(image,kernel)

plt.subplot(121)
plt.imshow(apertura,cmap='gray')

plt.subplot(122)
plt.imshow(cerradura,cmap='gray')

```
