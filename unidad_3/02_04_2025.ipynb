```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

path="../imagenes/pardoOriginal.png"
img=cv2.imread(path)     
mascara=np.array([[0,-1,1],[1,5,-1],[0,1,0]])
img_fin = cv2.filter2D(img,-1,mascara)

plt .subplot(121)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

plt.subplot(122)
plt.imshow(cv2.cvtColor(img_fin,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread(path)

kernerl5x5=np.ones((5,5),np.float32)/25 #suavizado
suavizado=cv2.filter2D(img,-1,kernerl5x5)#imagen suabizada

kernel7x7=np.ones((25,25),np.float32)/625
suavizado2=cv2.filter2D(img,-1,kernel7x7)

plt.figure(figsize=(10,10))

plt.subplot(141)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

plt.subplot(142)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

plt.subplot(143)
plt.imshow(cv2.cvtColor(suavizado,cv2.COLOR_BGR2RGB))

plt.subplot(144)
plt.imshow(cv2.cvtColor(suavizado2,cv2.COLOR_BGR2RGB))
```
