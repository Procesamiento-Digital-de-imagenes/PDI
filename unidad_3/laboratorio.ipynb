```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar las imgenes
imagen = cv2.imread('../imagenes/puntos.png', 0)
mascara = cv2.imread('../imagenes/pato.jpg', 0)

# Verificar si las imgenes se cargaron correctamente
if imagen is None or mascara is None:
    print("No se pudo cargar una o ambas imgenes. Verifica las rutas de los archivos.")
else:
    # Redimensionar la imagen para que coincida con el tamao de la mscara
    imagen_resized = cv2.resize(imagen, (mascara.shape[1], mascara.shape[0]))

    # Aplicar la imagen como mscara
    img_final = cv2.bitwise_and(mascara, mascara, mask=imagen_resized)

    # Mostrar la imagen final
    plt.imshow(img_final, cmap='gray')
    plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import measure

def contar2(img,valor1,valor2):
    img_gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    _,binarizada=cv2.threshold(img_gray,valor1,valor2,cv2.THRESH_BINARY_INV)
    contorno=measure.label(binarizada,connectivity=2)
    cont=np.max(contorno)
    return cont,binarizada


imagen=cv2.imread('../imagenes/colores.jpg')

rgb=cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)


img_sector_r=rgb[0:1000,0:300]
img_sector_o=rgb[0:1000,300:600]
img_sector_y=rgb[0:1000,570:870]
img_sector_g=rgb[0:1000,880:1090]
img_sector_b=rgb[0:1000,1100:1400]

cont_r,bin_r=contar2(img_sector_r,200,255)
cont_o,bin_o=contar2(img_sector_o,225,255)
cont_y,bin_y=contar2(img_sector_y,220,255)
cont_g,bin_g=contar2(img_sector_g,220,255)
cont_b,bin_b=contar2(img_sector_b,220,255)


plt.figure(figsize=(15,15))

plt.subplot(521)
plt.imshow(img_sector_r)

plt.subplot(522)
plt.imshow(bin_r,cmap='gray')
plt.title(f'Cantidad: {cont_r}')

plt.subplot(523)
plt.imshow(img_sector_o)

plt.subplot(524)
plt.imshow(bin_o,cmap='gray')
plt.title(f'Cantidad: {cont_o}')

plt.subplot(525)
plt.imshow(img_sector_y)

plt.subplot(526)
plt.imshow(bin_y,cmap='gray')
plt.title(f'Cantidad: {cont_y}')

plt.subplot(527)
plt.imshow(img_sector_g)

plt.subplot(528)
plt.imshow(bin_g,cmap='gray')
plt.title(f'Cantidad: {cont_g}')

plt.subplot(529)
plt.imshow(img_sector_b)

plt.subplot(5,2,10)
plt.imshow(bin_b,cmap='gray')
plt.title(f'Cantidad: {cont_b}')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_gray=cv2.imread('../imagenes/rostro.jpg',0)

canny = cv2.Canny(img_gray, 100,200)

plt.figure(figsize=(10,5))

plt.imshow(canny,cmap='gray')
```

EL filtro canny es una metodo por el cual mediante el uso de varias tecnicas como  suabizado ,umbralizacion,etc.
el resulto del filtro canny es la de la deteccion de los bordes de una imagen y al hacer esto nos permitiria ver los bordes de un rostro como se observa en la imagen y poder asi hacer la deteccion de rostros para nuestro proyecto.
