```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

def convolucion_media_horizontal(imagen,tamanioKernal):
    kernel=np.ones((1,tamanioKernal),np.float32)/tamanioKernal
    imgFin=cv2.filter2D(imagen,-1,kernel)
    return imgFin

path="../imagenes/dist.jpg"

imagen=cv2.imread(path)
gris=cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
imagen_rest=convolucion_media_horizontal(gris,9)

plt.figure(figsize=(10,5))

plt.subplot(131)
plt.imshow(cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB))

plt.subplot(132)
plt.imshow(gris,cmap='gray')

plt.subplot(133)
plt.imshow(imagen_rest,cmap='gray')
plt.show()
```

```python
def convolucion_media_horizontal(imagen,tamanioKernal):
    kernel=np.ones((1,tamanioKernal),np.float32)/tamanioKernal
    imgFin=cv2.filter2D(imagen,-1,kernel)
    return imgFin

#path="../imagenes/dist.jpg"
path="../imagenes/Distorcion.png"

imagen=cv2.imread(path)
rgb=cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
imgRest_r=convolucion_media_horizontal(rgb[:,:,0],99)
imgRest_g=convolucion_media_horizontal(rgb[:,:,1],99)
imgRest_b=convolucion_media_horizontal(rgb[:,:,2],99)

imgColor=np.stack([imgRest_r,imgRest_g,imgRest_b],axis=2)


plt.figure(figsize=(10,5))

plt.subplot(121)
plt.imshow(rgb)

plt.subplot(122)
plt.imshow(imgColor)
plt.show()

```
