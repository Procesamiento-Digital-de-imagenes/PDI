<h1 style='text-align:center'>Laplaciano</h1>

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt



img= cv2.imread('../imagenes/RELOJ.jpeg',cv2.IMREAD_GRAYSCALE)
laplacian = cv2.Laplacian(img, cv2.CV_64F)



plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(img,cmap='gray')

plt.subplot(122)
plt.imshow(laplacian,cmap='gray')
plt.show
```

<h1 style='text-align:center'>Laplaciano del Gaussiano</h1>

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt



img= cv2.imread('../imagenes/RELOJ.jpeg',cv2.IMREAD_GRAYSCALE)
blur = cv2.GaussianBlur(img, (3, 3), 0)

laplacian = cv2.Laplacian(blur, cv2.CV_64F)

plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(img,cmap='gray')

plt.subplot(122)
plt.imshow(laplacian,cmap='gray')
plt.show
```
