Cargado de imagenes

```python
import cv2
import matplotlib.pyplot as plt

path1="./img/img1.jpg"
path2="./img/img2.jpg"
path3="./img/img3.jpg"

img1=cv2.imread(path1)
img2=cv2.imread(path2)
img3=cv2.imread(path3)

plt.subplot(131)
plt.title("imagen1")
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))

plt.subplot(132)
plt.title("imagen2")
plt.imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB))

plt.subplot(133)
plt.title("imagen3")
plt.imshow(cv2.cvtColor(img3,cv2.COLOR_BGR2RGB))

plt.show()

```

escala de gris

canal escala de gary

```python
def escalaGris(img,axs):
    img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    if(not axs):
        plt.rcParams['figure.figsize']=[10,5]
        plt.subplot(141)
        plt.title("imgen Original")
        plt.imshow(img_rgb)

        plt.subplot(142)
        plt.imshow(img[:,:,0],cmap='gray')
        plt.title("Canal Rojo")

        plt.subplot(143)
        plt.imshow(img[:,:,1],cmap='gray')
        plt.title("Canal Verde")

        plt.subplot(144)
        plt.imshow(img[:,:,2],cmap='gray')
        plt.title("Canal Azul")
    else:
        fig,axs=plt.subplots(1,4,figsize=(20,5))

        axs[0].imshow(img_rgb)
        axs[0].set_title("Imagen original")

        fig_img=axs[1].imshow(img_rgb[...,0],cmap='gray')
        fig.colorbar(fig_img,ax=axs[1],shrink=0.5)
        axs[1].set_title('canal Rojo')

        img_g=axs[2].imshow(img_rgb[...,1],cmap='gray')
        fig.colorbar(fig_img,ax=axs[2],shrink=0.5)
        axs[2].set_title("Canal Verder")

        img_b=axs[3].imshow(img_rgb[...,2],cmap='gray')
        fig.colorbar(fig_img,ax=axs[3],shrink=0.5)
        axs[3].set_title("Canal Azul")
    
    
    plt.show()
```

mostrar color falso

```python
def ColorFalso(img):
    plt.rcParams['figure.figsize']=[10,5]
    b,g,r=cv2.split(img)
    plt.subplot(141)
    plt.title('Original')
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))


    plt.subplot(142)
    plt.title('Canal Rojo')
    plt.imshow(r,cmap="Reds")

    plt.subplot(143)
    plt.title('Canal Verde')
    plt.imshow(g,cmap="Greens")

    plt.subplot(144)
    plt.title('Canal Azul')
    plt.imshow(b,cmap="Blues")
```

Color real

```python
import numpy as np
def ColorReal(img):
    plt.rcParams['figure.figsize']=[10,5]
    img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    r=img[:,:,2]
    g=img[:,:,1]
    b=img[:,:,0]

    r_img=np.zeros_like(img_rgb)
    r_img[:,:,0]=r

    g_img=np.zeros_like(img_rgb)
    g_img[:,:,1]=g

    b_img=np.zeros_like(img_rgb)
    b_img[:,:,2]=b

    plt.subplot(1,4,1)
    plt.title("Imagen Original")
    plt.imshow(img_rgb)

    plt.subplot(1,4,2)
    plt.title("Canal Rojo")
    plt.imshow(r_img)

    plt.subplot(1,4,3)
    plt.title("Canal verde")
    plt.imshow(g_img)

    plt.subplot(1,4,4)
    plt.title("Canal Azul")
    plt.imshow(b_img)

```

funcion Mostrar imagenes

```python
def view_img(img,escala_gray,color_falso,color_real,plt_axs):
    if escala_gray:
        print("Escala de gris")
        escalaGris(img,plt_axs)
    if(color_real):
        print("Color Real")
        ColorReal(img)
    if color_falso:
        print("Falso Color")
        ColorFalso(img)
```
